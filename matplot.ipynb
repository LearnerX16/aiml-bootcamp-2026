```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

url="https://raw.githubusercontent.com/plotly/datasets/master/gapminderDataFiveYear.csv"

df=pd.read_csv(url)


print(df.head())
print(df.shape)
print(df.columns)
print(df.info())

```

```python
# Basic data analysis
print("Year range:", df["year"].min(), "-", df["year"].max())
print("Number of countries:", df["country"].nunique())
print("Number of continents:", df["continent"].nunique())
print("Average life expectancy:", df["lifeExp"].mean())
print("Minimum life expectancy:", df["lifeExp"].min())
print("Maximum life expectancy:", df["lifeExp"].max())

print(df.describe())
```

```python
#bar chart

# Average life expectancy by continent
avg_life = df.groupby("continent")["lifeExp"].mean()

# Bar chart
plt.figure(figsize=(8,5))
avg_life.plot(kind="bar", color=["skyblue", "orange", "green", "red", "purple"])

plt.title("Average Life Expectancy by Continent")
plt.xlabel("Continent")
plt.ylabel("Average Life Expectancy")
plt.xticks(rotation=0)
plt.show()
```

```python
# Select countries
selected_countries = ["India", "China", "United States"]

# Filter data
trend_df = df[df["country"].isin(selected_countries)]

# Create plot
plt.figure(figsize=(9, 5))

for country in selected_countries:
    country_df = trend_df[trend_df["country"] == country]

    plt.plot(
        country_df["year"],
        country_df["lifeExp"],
        marker="o",
        label=country
    )

# Formatting
plt.title("Life Expectancy Trend Over Time")
plt.xlabel("Year")
plt.ylabel("Life Expectancy")
plt.legend()
plt.grid(True)
plt.tight_layout()

# Show chart
plt.show()
```

```python
# Filter for 2007
df_2007 = df[df["year"] == 2007]

# Scatter plot
plt.figure(figsize=(8, 5))

plt.scatter(
    df_2007["gdpPercap"],
    df_2007["lifeExp"],
    alpha=0.7
)

plt.title("GDP per Capita vs Life Expectancy (2007)")
plt.xlabel("GDP per Capita")
plt.ylabel("Life Expectancy")
plt.grid(True)

plt.show()
```
