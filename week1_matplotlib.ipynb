```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 30, 25]

plt.plot(x, y)
plt.title("My First Matplotlib Graph")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.show()
```
